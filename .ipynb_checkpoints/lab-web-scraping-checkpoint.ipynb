```python
import requests           # To make HTTP requests
from bs4 import BeautifulSoup    # To parse HTML content
import pandas as pd     # To store and process the scraped data
import time       # To add delay between requests
```

```python
def scrape_books(min_rating=4, max_price=20):
    # Base URL of the site
    base_url = "http://books.toscrape.com/catalogue/page-{}.html"

    # Dictionary to convert text rating (e.g., "Four") into numeric (e.g., 4)
    # HTML contains class="star-rating Four", so we need to translate that
    rating_dict = {
        "One": 1,
        "Two": 2,
        "Three": 3,
        "Four": 4,
        "Five": 5
    }

    # List to collect all books
    books_data = []

    # Loop through all 50 pages of the site
    for page in range(1, 51):
        print(f"Scraping page {page}...")
        url = base_url.format(page)
        response = requests.get(url)

         # Check if the request was successful
        if response.status_code != 200:
            print(f"Failed to load page {page}, skipping.")
            continue

        # Parse the HTML of the current page
        soup = BeautifulSoup(response.content, "html.parser")
        books = soup.find_all("article", class_="product_pod")

        # Loop through all books on the current page
        for book in books:
            # Extract rating from  <p class="star-rating Five">
            rating_class = book.find("p", class_="star-rating")["class"][1]
            rating = rating_dict.get(rating_class, 0) # Convert text to number

            # Filter by minimum rating
            if rating < min_rating:
               continue

            # Get title and link to the detail page
            title = book.h3.a.get("title") or book.h3.a.get_text(strip=True)
            link = "http://books.toscrape.com/catalogue/" + book.h3.a["href"]

            try:
                # Request book detail page
                book_response = requests.get(link)
                if book_response.status_code != 200:
                    print(f"Failed to load book detail page: {link}")
                    continue

                book_soup = BeautifulSoup(book_response.content, "html.parser")

                # Extract price and convert to float
                price_text = book_soup.find("p", class_="price_color").text.strip()
                price = float(price_text[1:])  # Remove '£' and convert

                # Filter by max price
                if price > max_price:
                   continue

                # Get UPC (from product info table)
                upc = book_soup.find("th", string="UPC").find_next("td").text

                # Get availability
                availability = book_soup.find("p", class_="instock availability").get_text(strip=True)

                # Get description (if exists)
                description_tag = book_soup.find("meta", attrs={"name": "description"})
                description = description_tag["content"].strip() if description_tag else "No description"

                # Get genre (from breadcrumb navigation)
                genre = book_soup.find("ul", class_="breadcrumb").find_all("li")[2].text.strip()

                # Add book data to the list
                books_data.append({
                "UPC": upc,
                "Title": title,
                "Price (£)": price,
                "Rating": rating,
                "Genre": genre,
                "Availability": availability,
                "Description": description
                })

                # Be nice to the server
                time.sleep(0.5)

            except Exception as e:
                print(f"Error scraping book: {e}")
                continue

    # Convert list of books into a pandas DataFrame
    df = pd.DataFrame(books_data)
    return df
```

```python
# Run the function to scrape books with rating ≥ 4 and price ≤ 20
books_df = scrape_books(min_rating=4, max_price=20)
```
